## Criando Novas Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', ';')
dados.head()
```

```python
dados['Valor bruto'] = dados['Valor'] + dados['IPTU'] + dados['Condominio']
dados.head()
```

```python
dados['Valor m2'] = dados['Valor']/ dados['Area']
dados.head()
```

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
dados.head()
```

```python
dados['Valor Bruto m2'] = (dados['Valor bruto'] / dados['Area']).round(2)
dados.head(3)
```

```python
casa = ['Casa', 'Casa de Condominio', 'Casa de Vila']
# Usaremos apply para aplicar uma função para TODAS as linhas do dataframe
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados.head()
```
