## Criando Agrupamentos 

```python
import pandas as pd
```

```python
dados= pd.read_csv('dados/aluguel_residencial.csv', sep=';')
```

```python
dados.head(4)
```

```python
#bairros= ['Barra da Tijuca', 'Copacabana','Ipanema','Leblon','Botafogo','Tijuca']
#new_data=dados[dados['Bairro'].isin(bairros)]
```

```python
grupo_bairros= dados.groupby('Bairro')
#type(grupo_bairros)
```

```python
#grupo_bairros.groups
```

```python
for bairro, data in grupo_bairros:
    print('Bairro: ', bairro)
    print('Média: ', data.Valor.mean().round(4))
    print()
    #print(dados)
```

```python
grupo_bairros[['Valor']].mean().round(4).head()
```

# Estatística Descritivas

```python
grupo_bairros['Valor'].describe().round(2).head()
```

```python
grupo_bairros['Valor'].aggregate(['min', 'max']).rename(columns={'min':'Mínimo','max':'Máximo'})
```

```python
grupo_bairros['Valor'].aggregate(['min','max','sum']).rename(columns={'min':'Mínimo','max':'Máximo'})
```
