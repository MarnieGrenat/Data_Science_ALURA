```python
import polars as pl
```

```python
data=pl.read_csv('tips_atualizado.csv')
data.head(3)
```

```python
# Limpando e tratando os dados

# Modificando a nomenclatura das colunas para [PT-BR]
columns={
    'total_bill':'conta_total',
    'tip':'gorjeta',
    'dessert':'sobremesa',
    'day':'dia';
    'time':'horario',
    'size':'clientes'
}
data.columns=columns

# Transformando valores sobremesa
data[data['dia']== "No"] = False
data[data['dia']== "Yes"] = True

# Modificando valores horario
data.groupby(['time'])
data[data['horario']=='Dinner'] = 'Janta'
data[data['horario']=='Lunch'] = 'Almoco'

# Modificando valores dia
data.groupby(['day'])

#Checando dados faltantes
```
